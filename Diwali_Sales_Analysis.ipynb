# Exploratory Data Analysis

```python
ax= sns.countplot(x='Gender',data=df1)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
 df1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
```

```python
sales_gen = df1.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)

# Create the bar plot
ax = sns.barplot(x='Gender', y='Amount', data=sales_gen)

# Add annotations to the bars
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 5), 
                textcoords = 'offset points')


ax.set_title('Sales by Gender')
# Set labels and title
ax.set_xlabel('Gender')
ax.set_ylabel('Amount')

# Display the plot
plt.show()
```

<i>From the above graph we can see that most of the buyers are females and even the purchasing power of the females is greater than men</i>

# Age

```python
#count which age group does more shopping
sns_gen=sns.countplot(data=df1,x='Age Group', hue = 'Gender')
for bar in sns_gen.containers:
    sns_gen.bar_label(bar)
```

```python
#total amount vs age group
sns_age=df1.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sns_age,x='Age Group',y='Amount')
```

<i> from the above graphs we can see that most of the buyers are of age group between 26-35 yr female</i>

# States

```python
sns_state= df1.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sns_state,x='State',y='Orders')
```

```python
sns_state= df1.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sns_state,x='State',y='Amount')
```

*From the above graph we can see that most orders and sales/amount are from uttar pradesh,maharastra,karnataka., we can also see in top 10 states for orders 'kerela' is present but when it comes to sales/amount 'haryana' overtakes kerela. means although harayna has less orders but the amount of orders were high.* 

# Marital status

```python
ax=sns.countplot(data=df1,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,1)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Ms=df1.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_Ms,x='Marital_Status',y='Amount',hue='Gender')
```

*From the above graph we can see that most of the buyers are married(women) and they have high purchasing power*

# Occupation

```python
sns_jobs=sns.countplot(data=df1,x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bar in sns_jobs.containers:
    sns_jobs.bar_label(bar)
```

```python
sns_purchase_pow=df1.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sns_purchase_pow,x='Occupation',y="Amount")
```

*From the above graph we can see that most buyers are working in IT,Healthcare and  Aviation sector*

# Product Category

```python
sns_product=sns.countplot(data=df1,x='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
for bar in sns_product.containers:
    sns_product.bar_label(bar)
```

*Maximum number of order count is Clothing & apparel,food,electronics & gadgets*

```python
sns_product_amount =df1.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=sns_product_amount,x='Product_Category',y="Amount")
sns.set(rc={'figure.figsize':(20,5)})
```

*If you look carefully you can see that maximum number of orders is clothing whereas in sales the Food category has generated highest revenue. 
so, the most sold product is Food,clothing and electronics*

```python
sns_pid =df1.groupby(['Product_ID'],as_index=False)['product_Category'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sns_pid,x='Product_ID',y="Amount")
```
